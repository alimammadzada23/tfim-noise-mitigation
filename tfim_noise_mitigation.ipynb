```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager

from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel

from qiskit_ibm_runtime.fake_provider import FakeKyiv
from qiskit_ibm_runtime import QiskitRuntimeService, EstimatorV2 as Estimator

print("All imports OK ✅")
# ─── FILL THESE IN ───────────────────────────────────────────
MY_API_KEY = "--yours here__"
MY_CRN     = "--yours here__"
# ─────────────────────────────────────────────────────────────

# Save account to disk (only need to do this once per Colab session)
QiskitRuntimeService.save_account(
    channel="ibm_cloud",
    token=MY_API_KEY,
    instance=MY_CRN,
    overwrite=True
)

# Connect
service = QiskitRuntimeService(
    channel="ibm_cloud",
    token=MY_API_KEY,
    instance=MY_CRN
)

# List available real backends
backends = service.backends(operational=True, simulator=False)
print("✅ Connected! Available backends:")
for b in backends:
    print(f"   - {b.name}  |  {b.num_qubits} qubits")
# ─── Experiment Parameters ────────────────────────────────────
N_QUBITS  = 4       # number of spins in the chain
J         = 1.0     # ZZ coupling strength (ferromagnetic)
H_FIELD   = 0.1     # transverse field strength (spin flipper)
DT        = 0.2     # time slice per Trotter step
N_STEPS   = 8       # how many time steps we evolve through
# ─────────────────────────────────────────────────────────────

TIME_STEPS = list(range(1, N_STEPS + 1))
TIMES      = [t * DT for t in TIME_STEPS]

print(f"System:      {N_QUBITS} qubits | J={J} | h={H_FIELD}")
print(f"Time range:  t = {TIMES[0]:.1f} → {TIMES[-1]:.1f}")
print(f"Trotter steps: {N_STEPS}")
def build_trotter_circuit(n_qubits, n_steps, dt, J, h):
    """
    Trotterized TFIM time evolution circuit.
    Each step = ZZ interactions + X rotations.
    """
    qc = QuantumCircuit(n_qubits)

    # Initial state: all qubits in |+> (superposition of up and down)
    qc.barrier()

    for step in range(n_steps):
        # --- ZZ interaction (nearest-neighbor entanglement) ---
        for i in range(n_qubits - 1):
            qc.cx(i, i + 1)
            qc.rz(2 * J * dt, i + 1)
            qc.cx(i, i + 1)

        # --- Transverse field (X rotations on each qubit) ---
        for i in range(n_qubits):
            qc.rx(2 * h * dt, i)

        qc.barrier()

    return qc

# Build one circuit per time step
circuits = [build_trotter_circuit(N_QUBITS, t, DT, J, H_FIELD)
            for t in TIME_STEPS]

# Quick check
print(f"✅ Built {len(circuits)} circuits")
print(f"   Depth at t=1 : {circuits[0].depth()} gates")
print(f"   Depth at t=8 : {circuits[-1].depth()} gates")
print(f"\nCircuit for t=1 (one Trotter step):")
print(circuits[0].draw('text'))

#cell 6
def build_magnetization_observable(n_qubits):
    """
    Average Z observable: (Z0 + Z1 + Z2 + Z3) / n_qubits
    Each term is a Pauli string like 'IIIZ', 'IIZI', etc.
    """
    paulis = []
    for i in range(n_qubits):
        # Build string: I everywhere except position i which is Z
        pauli_str = 'I' * (n_qubits - 1 - i) + 'Z' + 'I' * i
        paulis.append((pauli_str, 1.0 / n_qubits))
    return SparsePauliOp.from_list(paulis)

observable = build_magnetization_observable(N_QUBITS)
print("Observable (average magnetization):")
print(observable)

#cell7
# Perfect noiseless simulator
ideal_sim = AerSimulator()
pm_ideal  = generate_preset_pass_manager(optimization_level=1, backend=ideal_sim)
circuits_ideal = pm_ideal.run(circuits)

# Run with Estimator
estimator_ideal = Estimator(mode=ideal_sim)
estimator_ideal.options.default_shots = 8192

job_ideal    = estimator_ideal.run([(c, observable) for c in circuits_ideal])
result_ideal = job_ideal.result()

ideal_values = [result_ideal[i].data.evs for i in range(len(circuits))]

print("✅ Ideal simulation done")
print("   Time   |   <Z>")
print("   -------|--------")
for t, v in zip(TIMES, ideal_values):
    bar = "█" * int(abs(v) * 20)
    print(f"   t={t:.1f}  |  {v:+.4f}  {bar}")
#cell8

# Load realistic noise from actual IBM hardware calibration
fake_backend = FakeKyiv()
noise_model  = NoiseModel.from_backend(fake_backend)
noisy_sim    = AerSimulator(noise_model=noise_model)

# Transpile for the fake backend's native gate set
pm_noisy       = generate_preset_pass_manager(optimization_level=1, backend=fake_backend)
circuits_noisy = pm_noisy.run(circuits)

# ✅ FIX: map the 4-qubit observable onto the physical qubit layout
# After transpilation, our 4 qubits live at specific positions inside 127 qubits
# apply_layout() rewrites the observable to match that physical mapping
observables_noisy = [observable.apply_layout(c.layout) for c in circuits_noisy]

estimator_noisy = Estimator(mode=noisy_sim)
estimator_noisy.options.default_shots = 8192

job_noisy    = estimator_noisy.run(
    [(c, obs) for c, obs in zip(circuits_noisy, observables_noisy)]
)
result_noisy = job_noisy.result()

noisy_values = [result_noisy[i].data.evs for i in range(len(circuits))]

# How much did noise damage the result?
raw_errors = [abs(ideal_values[i] - noisy_values[i]) for i in range(len(circuits))]

print("✅ Noisy simulation done")
print(f"   Max error due to noise: {max(raw_errors):.4f}")
print(f"   Avg error due to noise: {np.mean(raw_errors):.4f}")
print("\n   Time   |   Ideal   |   Noisy   |   Error")
print("   -------|-----------|-----------|--------")
for t, iv, nv, e in zip(TIMES, ideal_values, noisy_values, raw_errors):
    flag = " ⚠️" if e > 0.05 else ""
    print(f"   t={t:.1f}  |  {iv:+.4f}  |  {nv:+.4f}  |  {e:.4f}{flag}")

#cell9

def fold_circuit(circuit, scale_factor):
    if scale_factor == 1:
        return circuit

    folded = QuantumCircuit(circuit.num_qubits)
    n_extra_pairs = (scale_factor - 1) // 2

    for instruction in circuit.data:
        gate   = instruction.operation
        qubits = instruction.qubits

        if gate.name == 'barrier':
            folded.barrier()
            continue

        folded.append(gate, qubits)

        for _ in range(n_extra_pairs):
            try:
                folded.append(gate.inverse(), qubits)
            except Exception:
                folded.append(gate, qubits)
            folded.append(gate, qubits)

    return folded


# ─── Run at 3 noise scale factors ────────────────────────────
SCALE_FACTORS = [1, 3, 5]
zne_raw = {}

for scale in SCALE_FACTORS:
    print(f"Running scale factor {scale}× ...", end=" ")

    folded_circuits = [fold_circuit(qc, scale) for qc in circuits_noisy]

    est = Estimator(mode=noisy_sim)
    est.options.default_shots = 8192

    # ✅ FIX: use observables_noisy (already mapped to 127-qubit layout)
    job = est.run([(c, obs) for c, obs in zip(folded_circuits, observables_noisy)])
    res = job.result()

    # ✅ NEW - wrap with float() to convert numpy scalar to Python float
    zne_raw[scale] = [float(np.array(res[i].data.evs).flatten()[0]) for i in range(len(circuits))]
    print(f"done → {[round(v, 3) for v in zne_raw[scale]]}")


# ─── Extrapolate to zero noise ────────────────────────────────
def linear_extrapolate_to_zero(scale_factors, values):
    coeffs = np.polyfit(scale_factors, values, deg=1)
    return float(np.polyval(coeffs, 0))

# ✅ Safe extraction for mitigated values
mitigated_values = []
for i in range(len(circuits)):
    vals    = [zne_raw[s][i] for s in SCALE_FACTORS]
    zne_val = linear_extrapolate_to_zero(SCALE_FACTORS, vals)
    mitigated_values.append(float(zne_val))


# ─── Report improvement ───────────────────────────────────────
mitigated_errors = [abs(ideal_values[i] - mitigated_values[i]) for i in range(len(circuits))]

print("\n✅ ZNE mitigation done")
print(f"   Max error BEFORE ZNE : {max(raw_errors):.4f}")
print(f"   Max error AFTER  ZNE : {max(mitigated_errors):.4f}")
reduction = (1 - max(mitigated_errors) / max(raw_errors)) * 100
print(f"   Error reduction      : {reduction:.1f}% 🎯")

print("\n   Time   |   Ideal   |   Noisy   |  Mitigated")
print("   -------|-----------|-----------|----------")
for t, iv, nv, mv in zip(TIMES, ideal_values, noisy_values, mitigated_values):
    print(f"   t={t:.1f}  |  {iv:+.4f}  |  {nv:+.4f}  |  {mv:+.4f}")


#cell10

fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle(
    'TFIM Hamiltonian Simulation  |  4 Qubits  |  ZNE Error Mitigation',
    fontsize=13, fontweight='bold', y=1.02
)

# ── Left panel: three curves ──────────────────────────────────
ax = axes[0]
ax.plot(TIMES, ideal_values,     'o-',  color='#2ECC71', lw=2.5, ms=8, label='✅ Ideal (noiseless)',   zorder=3)
ax.plot(TIMES, noisy_values,     's--', color='#E74C3C', lw=2.0, ms=7, label='❌ Noisy (FakeKyiv)',    zorder=2)
ax.plot(TIMES, mitigated_values, '^-',  color='#3498DB', lw=2.5, ms=8, label='🔵 ZNE Mitigated',       zorder=3)

ax.set_xlabel('Time  t', fontsize=12)
ax.set_ylabel('Average Magnetization  ⟨Z⟩', fontsize=12)
ax.set_title('Magnetization vs Time', fontsize=11)
ax.legend(fontsize=10)
ax.grid(True, alpha=0.3)
ax.axhline(0, color='gray', lw=0.8, ls=':')
ax.set_ylim(-1.1, 1.1)

# ── Right panel: error comparison bar chart ───────────────────
ax2   = axes[1]
x     = np.arange(len(TIMES))
width = 0.35

ax2.bar(x - width/2, raw_errors,       width, label='Noisy error',   color='#E74C3C', alpha=0.85)
ax2.bar(x + width/2, mitigated_errors, width, label='ZNE error',     color='#3498DB', alpha=0.85)

ax2.set_xlabel('Time step', fontsize=12)
ax2.set_ylabel('|Ideal − Result|', fontsize=12)
ax2.set_title(f'Error Reduction via ZNE  ({reduction:.0f}% improvement)', fontsize=11)
ax2.set_xticks(x)
ax2.set_xticklabels([f't={t:.1f}' for t in TIMES], rotation=40)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.savefig('tfim_zne_results.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plot saved as tfim_zne_results.png")

#cell11

# ============================================================
# LAYERED ERROR MITIGATION ON REAL IBM HARDWARE
# Three mitigations: DD + TREX + ZNE combined
# ============================================================

from qiskit_ibm_runtime import EstimatorOptions

real_backend = service.least_busy(
    operational=True,
    simulator=False,
    min_num_qubits=4
)
print(f"Running on: {real_backend.name}")

# Transpile for real hardware
pm_real       = generate_preset_pass_manager(optimization_level=3, backend=real_backend)
circuits_real = pm_real.run(circuits)
observables_real = [observable.apply_layout(c.layout) for c in circuits_real]

# ─── Build 3 separate estimators with different mitigation levels ───

# LEVEL 0 — No mitigation (raw hardware, worst case)
opts_raw = EstimatorOptions()
opts_raw.resilience_level         = 0
opts_raw.default_shots            = 4096

# LEVEL 1 — Readout mitigation only (fixes measurement bias)
# TREX: randomly twirls measurement basis to cancel systematic readout errors
opts_trex = EstimatorOptions()
opts_trex.resilience_level        = 1   # enables TREX automatically
opts_trex.default_shots           = 4096

# LEVEL 2 — Full mitigation: TREX + ZNE + Dynamical Decoupling
opts_full = EstimatorOptions()
opts_full.resilience_level        = 2   # enables ZNE + TREX
opts_full.default_shots           = 4096
# Dynamical Decoupling: inserts X-X pulse pairs during idle qubit times
# prevents decoherence while other qubits are busy with gates
opts_full.dynamical_decoupling.enable       = True
opts_full.dynamical_decoupling.sequence_type = "XY4"  # best general-purpose DD sequence

# ─── Submit all 3 jobs ─────────────────────────────────────────────
print("\nSubmitting 3 jobs to real hardware...")

est_raw  = Estimator(mode=real_backend, options=opts_raw)
est_trex = Estimator(mode=real_backend, options=opts_trex)
est_full = Estimator(mode=real_backend, options=opts_full)

job_raw  = est_raw.run([(c, obs) for c, obs in zip(circuits_real, observables_real)])
job_trex = est_trex.run([(c, obs) for c, obs in zip(circuits_real, observables_real)])
job_full = est_full.run([(c, obs) for c, obs in zip(circuits_real, observables_real)])

print(f"Job RAW  ID : {job_raw.job_id()}")
print(f"Job TREX ID : {job_trex.job_id()}")
print(f"Job FULL ID : {job_full.job_id()}")
print("\n⏳ Waiting for all 3 jobs... (may take several minutes)")

# ─── Collect results ───────────────────────────────────────────────
def get_values(job):
    res = job.result()
    return [float(np.array(res[i].data.evs).flatten()[0]) for i in range(len(circuits))]

values_raw  = get_values(job_raw)
values_trex = get_values(job_trex)
values_full = get_values(job_full)

# ─── Print comparison ──────────────────────────────────────────────
print("\n✅ All jobs done!")
print(f"\n{'Time':>6} | {'Ideal':>8} | {'Raw HW':>8} | {'TREX':>8} | {'Full Mit':>8} | {'Improvement'}")
print("-" * 75)

for i, t in enumerate(TIMES):
    iv   = ideal_values[i]
    raw  = values_raw[i]
    trex = values_trex[i]
    full = values_full[i]

    err_raw  = abs(iv - raw)
    err_full = abs(iv - full)
    improved = "✓ better" if err_full < err_raw else "✗ worse"

    print(f"t={t:.1f}  | {iv:>+8.4f} | {raw:>+8.4f} | {trex:>+8.4f} | {full:>+8.4f} | {improved}")

#cell12

# ============================================================
# LAYERED ERROR MITIGATION ON REAL IBM HARDWARE
# Three mitigations: DD + TREX + ZNE combined
# ============================================================

from qiskit_ibm_runtime import EstimatorOptions

real_backend = service.least_busy(
    operational=True,
    simulator=False,
    min_num_qubits=4
)
print(f"Running on: {real_backend.name}")

# Transpile for real hardware
pm_real       = generate_preset_pass_manager(optimization_level=3, backend=real_backend)
circuits_real = pm_real.run(circuits)
observables_real = [observable.apply_layout(c.layout) for c in circuits_real]

# ─── Build 3 separate estimators with different mitigation levels ───

# LEVEL 0 — No mitigation (raw hardware, worst case)
opts_raw = EstimatorOptions()
opts_raw.resilience_level         = 0
opts_raw.default_shots            = 4096

# LEVEL 1 — Readout mitigation only (fixes measurement bias)
# TREX: randomly twirls measurement basis to cancel systematic readout errors
opts_trex = EstimatorOptions()
opts_trex.resilience_level        = 1   # enables TREX automatically
opts_trex.default_shots           = 4096

# LEVEL 2 — Full mitigation: TREX + ZNE + Dynamical Decoupling
opts_full = EstimatorOptions()
opts_full.resilience_level        = 2   # enables ZNE + TREX
opts_full.default_shots           = 4096
# Dynamical Decoupling: inserts X-X pulse pairs during idle qubit times
# prevents decoherence while other qubits are busy with gates
opts_full.dynamical_decoupling.enable       = True
opts_full.dynamical_decoupling.sequence_type = "XY4"  # best general-purpose DD sequence

# ─── Submit all 3 jobs ─────────────────────────────────────────────
print("\nSubmitting 3 jobs to real hardware...")

est_raw  = Estimator(mode=real_backend, options=opts_raw)
est_trex = Estimator(mode=real_backend, options=opts_trex)
est_full = Estimator(mode=real_backend, options=opts_full)

job_raw  = est_raw.run([(c, obs) for c, obs in zip(circuits_real, observables_real)])
job_trex = est_trex.run([(c, obs) for c, obs in zip(circuits_real, observables_real)])
job_full = est_full.run([(c, obs) for c, obs in zip(circuits_real, observables_real)])

print(f"Job RAW  ID : {job_raw.job_id()}")
print(f"Job TREX ID : {job_trex.job_id()}")
print(f"Job FULL ID : {job_full.job_id()}")
print("\n⏳ Waiting for all 3 jobs... (may take several minutes)")

# ─── Collect results ───────────────────────────────────────────────
def get_values(job):
    res = job.result()
    return [float(np.array(res[i].data.evs).flatten()[0]) for i in range(len(circuits))]

values_raw  = get_values(job_raw)
values_trex = get_values(job_trex)
values_full = get_values(job_full)

# ─── Print comparison ──────────────────────────────────────────────
print("\n✅ All jobs done!")
print(f"\n{'Time':>6} | {'Ideal':>8} | {'Raw HW':>8} | {'TREX':>8} | {'Full Mit':>8} | {'Improvement'}")
print("-" * 75)

for i, t in enumerate(TIMES):
    iv   = ideal_values[i]
    raw  = values_raw[i]
    trex = values_trex[i]
    full = values_full[i]

    err_raw  = abs(iv - raw)
    err_full = abs(iv - full)
    improved = "✓ better" if err_full < err_raw else "✗ worse"

    print(f"t={t:.1f}  | {iv:>+8.4f} | {raw:>+8.4f} | {trex:>+8.4f} | {full:>+8.4f} | {improved}")
```
